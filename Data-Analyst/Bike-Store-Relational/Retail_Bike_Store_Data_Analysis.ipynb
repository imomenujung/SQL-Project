Checking Misssing Value

```python
query = """
SELECT
    strftime('%Y', order_date) as order_year,
    shipped_date,
    COUNT(1) as num_order
FROM orders
WHERE
    shipped_date IS NULL
GROUP BY 1
"""

df = pd.read_sql_query(query, connection)
df
```

Based on the analysis results, it can be seen that there are still missing values on the Shipped Date.
This can happen because the information has not been updated. Therefore, we will update the value to "Need Update" so that companies can know and make decisions in the future.



```python
query = """
SELECT
    order_id,
    order_date,
    COALESCE(shipped_date, 'Need Update') AS shipped_date
FROM orders
/*By using COALESCE, we can change the NULL values into anything we describe*/
WHERE
    COALESCE(shipped_date, 'Need Update') = 'Need Update'
"""
df = pd.read_sql_query(query, connection)
df
```

```python
# if the question defines "each month" by total sales for each month by separating the years
query = """
SELECT
    strftime('%Y', order_date) as year,
    strftime('%m', order_date) as month,
    SUM(quantity) as unit_sold
FROM orders o
    INNER JOIN
    order_items oi
        ON o.order_id = oi.order_id
GROUP BY 1,2
        """
df = pd.read_sql_query(query, connection)
df
```

If we look at the previous monthly sales analysis, we may identify a problem of customer order patterns after June 2018. Therefore, we need to investigate this issue.

First, we want to write a query to identify customers who placed orders in each month for the last six months.

```python
query = """
SELECT
    strftime('%Y', o.order_date) as year,
    strftime('%m', o.order_date) as month,
    c.customer_id
FROM customers c
JOIN orders o
    ON o.customer_id = c.customer_id
WHERE
    strftime('%Y', o.order_date) = '2018' AND
    strftime('%m', o.order_date) >= '06'
ORDER BY 2 ASC, 3 ASC
"""
df = pd.read_sql_query(query, connection)
df
```

We can see that there is a problem of sales within the last 6 months, which shows very low sales. This can also be seen in the previous monthly sales analysis which also confirms this extreme fall of sales.

Therefore, in the next analysis, we will try to segment the customers using Dillon Myrick analysis on his notebook (1), to try to understand the patterns better. Here, I update Myrick's definition of recent_buyer to be those who made purchase for the latest year, because using 90 days before like the one Myrick used is not relevant based on the previous revelation.

```python
query = """
WITH customer_stats AS (
    SELECT
        customer_id,
        SUM(quantity * list_price * (1 - discount)) AS total_spending,
        COUNT(DISTINCT o.order_id) AS total_orders,
        julianday('2018-12-31') - julianday(MAX(order_date)) AS days_since_last_purchase
    FROM
        orders o
    INNER JOIN
        order_items oi
    ON
        o.order_id = oi.order_id
    GROUP BY
        1
)

SELECT
    customer_id,
    CASE WHEN total_orders > 1 THEN 'repeat buyer'
         ELSE 'one-time buyer'
         END AS purchase_frequency,
    CASE WHEN days_since_last_purchase < 360 THEN 'recent buyer'
         ELSE 'not recent buyer'
         END AS purchase_recency,
    CASE WHEN total_spending/(SELECT MAX(total_spending) FROM customer_stats) >= .65 THEN 'big spender'
         WHEN total_spending/(SELECT MAX(total_spending) FROM customer_stats) <= .30 THEN 'low spender'
         ELSE 'average spender'
         END AS buying_power
FROM
    customer_stats
"""

df = pd.read_sql_query(query, connection)
df
```

calculate the month-over-month sales growth for the past year.

```python
query = """
SELECT
    strftime('%Y', order_date) as year,
    strftime('%m', order_date) as month,
    SUM(quantity) as unit_sold,
    SUM(quantity * oi.list_price * (1 - oi.discount)) AS total_sales
FROM orders o
    INNER JOIN
    order_items oi
        ON o.order_id = oi.order_id
GROUP BY 1,2
        """
df = pd.read_sql_query(query, connection)
df
```

From the results above, it can be said that the data after May 2018 is messed up so the data used is only from 2016 and 2017.

```python
query_2016 = """
WITH sales_data AS (
    SELECT
        strftime('%Y', o.order_date) as year,
        strftime('%m', o.order_date) as month,
        SUM(oi.quantity * oi.list_price * (1 - oi.discount)) as sales_amount,
        LAG(SUM(oi.quantity * oi.list_price * (1 - oi.discount)), 1) OVER (ORDER BY strftime('%m', o.order_date)) AS prev_sales_amount
    FROM
        orders o
    JOIN
        order_items oi ON o.order_id = oi.order_id
    WHERE year = '2016'

    GROUP BY
        strftime('%m', o.order_date)
)

SELECT
    month,
    year,
    sales_amount,
    prev_sales_amount,
    CASE
        WHEN prev_sales_amount IS NULL THEN NULL  -- Handle the first month case
        ELSE (1.0 * (sales_amount - prev_sales_amount) / prev_sales_amount) * 100.0
    END AS sales_growth_percent
FROM
    sales_data
GROUP BY month
ORDER BY
    month;
    """
df_2016 = pd.read_sql_query(query_2016, connection)
df_2016
```

```python
query_2017 = """
WITH sales_data AS (
    SELECT
        strftime('%Y', o.order_date) as year,
        strftime('%m', o.order_date) as month,
        SUM(oi.quantity * oi.list_price * (1 - oi.discount)) as sales_amount,
        LAG(SUM(oi.quantity * oi.list_price * (1 - oi.discount)), 1) OVER (ORDER BY strftime('%m', o.order_date)) AS prev_sales_amount
    FROM
        orders o
    JOIN
        order_items oi ON o.order_id = oi.order_id
    WHERE year = '2017'

    GROUP BY
        strftime('%m', o.order_date)
)

SELECT
    month,
    year,
    sales_amount,
    prev_sales_amount,
    CASE
        WHEN prev_sales_amount IS NULL THEN NULL  -- Handle the first month case
        ELSE (1.0 * (sales_amount - prev_sales_amount) / prev_sales_amount) * 100.0
    END AS sales_growth_percent
FROM
    sales_data
GROUP BY month
ORDER BY
    month;
    """
df_2017 = pd.read_sql_query(query_2017, connection)
df_2017
```

```python
import plotly.express as px

# Sales growth over time (2016)
df_2016.sort_values(by='month', inplace=True)
fig_2016 = px.line(df_2016, x='month', y='sales_growth_percent', title='Sales Growth Over Time (2016)')
fig_2016.update_layout(xaxis_title='Month', yaxis_title='Total Sales')
fig_2016.show()

# Sales growth over time (2017)
df_2017.sort_values(by='month', inplace=True)
fig_2017 = px.line(df_2017, x='month', y='sales_growth_percent', title='Sales Growth Over Time (2017)')
fig_2017.update_layout(xaxis_title='Month', yaxis_title='Total Sales')
fig_2017.show()
```
